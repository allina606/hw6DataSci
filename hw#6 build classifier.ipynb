```python
# open the file
import pandas as pd
data = pd.read_csv('hw6.data.csv', header=None)

data 
```

```python
# organize into 70% training and 30% testing
trainingDF = data.iloc[0:6999]
testingDF = data.iloc[7000:10000]

#making sure  dataframe has correct vals 
# print(trainingDF.count())
# print(testingDF.count())

trainingDF[10]
```

```python
#construct multiple linear regression (ordinary least squares regression) classifier
import numpy as np
import warnings
warnings.filterwarnings('ignore')

#building model
# note: underscore means private func
class Model(object):
    def __init__(self):
        self.coefficients = []
        
    # take this out later
    def _reshape_input(self, x):
        return x.reshape(-1,1)
    
    # bias term = adding vector of 1s to x
    def _add_bias_term(self, x):
        ones = np.ones(shape=x.shape[0]).reshape(-1,1)
        return np.concatenate( ( ones, x ),1 )
    
    def fit(self, x, output):
        if len(x.shape) == 1: 
            x=self._reshape_input(x)
            
        x=self._add_bias_term(x)
        self.coefficients = np.linalg.inv(x.transpose().dot(x)).dot(x.transpose()).dot(output)
        
    def predict(self, val):
        b0 = self.coefficients[0]
        other_betas = self.coefficients[1:]
        prediction = b0
        
        for xi, bi in zip(val, other_betas): prediction += (bi*xi)
        
        return prediction;
```

```python
# train model on training data

# construct feature matrix and target vector 
feature = trainingDF.drop(10, axis=1).values
target = trainingDF[10].values

model = Model()

model.fit(feature, target)

model.coefficients
```

```python
# report accuracy using test set
featureTest = testingDF.drop(10,axis=1).values
targetTest = testingDF[10].values

targetIdx = 0;
print('predicted= \t\tactual = ')
for idx in featureTest: 
    print(model.predict(idx), '\t', targetTest[targetIdx])
    targetIdx = targetIdx+1
```
